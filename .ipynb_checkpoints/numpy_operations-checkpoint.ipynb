```python
import numpy as np
import time
```

```python
#excution performance
size = 1_000_000

py_list = list(range(size))
start = time.time()
sq_list = [i**2 for i in py_list]
end = time.time()
print(f"py time : {end  -start }sec " )

np_arr = np.array(py_list)
start = time.time()
sq_list = np_arr **2
end = time.time()
print(f"np time : {end  -start }sec " )
```
