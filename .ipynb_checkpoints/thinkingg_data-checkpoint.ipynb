```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        data = json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
   text_to_num = {"one" : 1,"two":2,"three":3,"four":4,"five":5}
   cleaned_data = []
   unique_data = set()
   for user in data:
       raw_rating=  str(user["rating"]).strip().lower()
       if(raw_rating in text_to_num):
           raw_rating = text_to_num[raw_rating]
           user["rating"] = raw_rating

        
       raw_age = user.get("age")
       if(raw_age == None):
           user["age"] = None
        
            #duplicate data
       if(user["name"] in unique_data):
          continue
       unique_data.add(user["name"])
       cleaned_data.append(user)
        
   return cleaned_data
```

```python
data = clean_data(data)
```

```python
def get_insightss(data):
    total_rating = 0
    #avg rating
    for user in data:
        total_rating += float( user["rating"])
    print(f"avg rating : {total_rating/len(data)*100}%")
    #poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) <3):
            poor_rating+=1
    print(f"poor rating : {poor_rating/len(data)*100}%")
          
get_insightss(data)
```
